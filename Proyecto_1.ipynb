```python
#Emily Melendez Guerrero

import pandas as pd #Importamos pandas para que pueda leer el achivo de exel

class extraer_data_frame():
  "Esta es la sintaxis correcta para extraer nuestro data frame y nuestro data frame es lo mas importante para comenzar a generar la ruta"
  def leer_archivo(data_excel):
    datos = pd.read_excel(data_excel)
    return datos

def extraer_columnas(datos):
    "Nos va a regresas las columnas importantes para la elaboracion del proyecto"

    actividades = datos["Actividad"]
    precedentes = datos["Precedentes"]
    duracion = datos["Duracion"]
    return (actividades, precedentes, duracion) #Queremos estos valores de regreso

def gererame_la_ruta():
  "La primera parte de esto es que metemos nuestro archivo y comienza leyendo lo que nos interesa para asi poder regresarlo y que generemos la ruta"

  arch = '/content/Creacion_de_la_carrera_de_mat_ap.xlsx' #Es el archivo de ejemplo pero se podria meter para cualquier otro archivo en formato exel
  data = leer_archivo(arch)
  #print(data)

  act, prec, dur = extraer_columnas(data) #Lo mismo que arriba son las columnas que nos interesan
  #print(dur) son solo print para checar
  #print(prec) solo para checar

  "En esta parte lo que hace es que con base a lo anterior y a la lectura del Data_Frame comienza a generar mi ruta con las condiciones que le voy pidiendo"

  tupla = []
  espera = []
  precedentes = []

  for i in range(len(act)):
    if isinstance(prec[i], str):
        tupla_prec = tuple(prec[i].split(",")) #Tenemos los casos de los que tienen doble precedente y esta parte nos ayuda a separarlos
        for antecedente in tupla_prec:
            if int(antecedente) < int(act[i]): #Se fija si el antecedente es menor que la actividad que lo meta a la tupla
                tupla.append((act[i], antecedente, dur[i]))
            else:
                espera.append((act[i], antecedente, dur[i])) #si no pasa que se meta a la de espera
    else:
        tupla.append((act[i], prec[i], dur[i]))

    #print(act[i])
    #print(espera)


    if len(espera)>0 and act[i] == int(espera[0][1]):
      tupla.append((espera[0][0], espera [0][1], espera [0][2]))
      espera.pop (0)

  #print("ruta ", tupla)
  return tupla #ESTA PARTE GENERA MI RUTA

class Directed_graph:

    def init(self):
        self.nodos = []
        self.arcos = []

    def len_arcos(self):
        return len(self.arcos)

    def len_nodos(self):
        return len(self.nodos)

    def antecesores(self, nodo):
        if nodo not in self.nodos:
            print(nodo, "NO PERTENECE AL CONJUNTO DE NODOS")
            return []
        else:
            ant = []
            for x in self.arcos:
                if nodo == x[-1]:
                    ant.append(x[0])
            return ant

    def sucesor(self, nodo):
        if nodo not in self.arcos:
            print(nodo, "No pertenece al conjunto de nodos")
            return []
        else:
            ant = []
            for x in self.arcos:
                if nodo == x[0]:
                    ant.append(x[-1])
            return ant

    def add_vertex(self, nodo):
        self.nodos.append(nodo)

    def add_arc(self, arco):
        if isinstance(arco, tuple) and len(arco) == 3:
            if arco[0] in self.nodos and arco[1] in self.nodos:
                self.arcos.append(arco)
            else:
                print("LAS ENTRADAS NO PERTENECEN A LOS NODOS")
        else:
            print("EL ARCO NO ES VALIDO")

dic = {1: [], 2: [1], 3: [2], 4: [3], 5: [4], 6: [4], 7: [4], 8: [4], 9: [4],
       10: [5], 11: [6], 12: [7], 13: [8], 14: [11], 15: [10, 14], 18: [15],
       19: [18], 16: [12, 13, 18], 17: [9, 16], 20: [17, 19], 21: [20]} #Esto lo obtenemos un diccionario com el orden de la  actividad y precedente

subsecuentes = {}
arcos = {}
nodos = [0, 1]
done = [[], []]
ind = 2
arrows = []
arcos = {}
for a, ps in dic.items():     # a = actividad, ps = precedentes
  subsecuentes[a] = []
  for p in ps:
    subsecuentes[p].append(a)
#print(subsecuentes)

for k, v in dic.items():
  if v == []:
    nodos.append(ind)
    arrows.append((0, ind))
    done.append([k])
    arcos[k] = (0, ind)
    ind +=1
    #print(arrows)
  elif not v in done:
    separated = [[elem] for elem in v]
    for elem in separated:
      arrows.append((done.index(elem), ind))
    done.append(v)
    nodos.append(ind)
    ind += 1
    #print(arrows)
    if subsecuentes[k] == []:
      arrows.append((done.index(v), 1))
      arcos[k] = (done.index(v, 1))
    else:
      nodos.append(ind)
      arrows.append((done.index(v), ind))
      done.append([k])
      arcos[k] = (done.index(v), ind)
      ind += 1
  else:
      if subsecuentes[k] == []:
        arrows.append((done.index(v), 1))
        arcos[k] = (done.index(v), 1)
      else:
        nodos.append(ind)
        arrows.append((done.index(v), ind))
        done.append([k])
        arcos[k] = (done.index(v), ind)
        ind +=1

#print(nodos) #De aqui obtengo mis nodos
#print(arrows) #De aqui obtengo mis arcos

l=[(1, (0,), 3), (2, (1,), 15), (3, (2,), 10), (4, (3,), 2), (5, (4,), 8),
 (6, (4,), 10), (7, (4,), 10), (8, (4,), 8), (9, (4,), 5), (10, (5,), 20),
  (11, (6,), 10), (12, (7,), 8), (13, (8,), 6), (14, (11,), 10),
   (15, (10, 14), 20), (18, (15,), 3), (16, (12, 13, 18), 15), (17,
    (9, 16), 10), (19, (18,), 8), (20, (17, 19), 8), (21, (20,), 8)]  #Esta parte se obtuvo de generame la ruta y solo lo estamos recuperando aqui y el orden es actividad, precedente y duracion



def checar_precedentes_repetidos(tupla):
    "Esto encarga de checar si alguno de los precedentes esta repetido que son los que nos causan problemas y ya con esto podemos decir cuales son los que se repiten "

    precedentes_repetidos = []
    tuplas_con_precedentes_iguales = []
    precedentes = set() #Estos es para ver checar que una funcion no tenga duplicados

    for i in tupla:
        max_1 = max(i[1])
        if max_1 in precedentes:
            precedentes_repetidos.append(max_1)
            tuplas_con_precedentes_iguales.append(i)
        else:
            precedentes.add(max_1)

    return tuplas_con_precedentes_iguales, precedentes_repetidos

#print("Repetidos", checar_precedentes_repetidos(l))

def fecha_mas_proxima(tupla):
    "Nos calcula las fechas mas proximas de cada una de las actividades"
    FMP = {} #Inicializamos un diccionario
    for actividad, precedentes, duracion in tupla:
        FMP[actividad] = 0  #Aqui inicializamos en 0 para que las calcule de forma correcta
        for precedente in precedentes:
            if precedente not in FMP:
                FMP[precedente] = 0

    for actividad, precedentes, duracion in tupla:
        if precedentes:
            FMP[actividad] = max(FMP[actividad], max(FMP[precedente] for precedente in precedentes) + duracion) #FORMULA DE JOSE
        else:
            FMP[actividad] = max(FMP[actividad], duracion)
    return FMP

def fechas_mas_lejanas(tupla, FMP):
    "Nos calcula las fechas mas lejanas de cada una de las actividades"
    FML = {act: max(FMP.values()) for act, _, _ in tupla}
    for act in sorted(FMP, key=FMP.get, reverse=True):
        for actividad, precedentes, duracion in tupla:
            if actividad == act:
                if precedentes:
                    for precedente in precedentes:
                        FML[precedente] = min(FML[precedente], FML[actividad] - duracion) #FORMULA DE JOSE

    return FML

def calcular_holguras(tupla, FMP, FML):
    "Nos calcula las holguras de cada una de las actividades, esto es lo mas importante porque cada una de las que su calculo de 0 significa que es una actividad critica"
    holguras = {} #Inicializamos un diccionario
    for actividad, precedentes, duracion in tupla:
        if actividad in FML and actividad in FMP:
            # La holgura es la diferencia entre la fecha más lejana y la fecha más próxima
            holgura_total = FML[actividad] - FMP[actividad] #FORMULA DE JOSE
            holguras[actividad] = holgura_total
    return holguras

def main():
  l = [(1, (), 3), (2, (1,), 15), (3, (2,), 10), (4, (3,), 2), (5, (4,), 8),
   (6, (4,), 10), (7, (4,), 10), (8, (4,), 8), (9, (4,), 5), (10, (5,), 20),
    (11, (6,), 10), (12, (7,), 8), (13, (8,), 6), (14, (11,), 10),
     (15, (10, 14), 20), (18, (15,), 3), (16, (12, 13, 18), 15),
      (17, (9, 16), 10), (19, (18,), 8), (20, (17, 19), 8), (21, (20,), 8)]

  FMP = fecha_mas_proxima(l)
  FML = fechas_mas_lejanas(l, FMP)
  holguras = calcular_holguras(l, FMP, FML)

  #print("FMP:", FMP)
  #print("FML: ", FML) #NO QUISE QUITAR NINGUN PRINT PARA IR CHECANDO EL CODIGO
  #rint("HOLGURAS: ", holguras)

main()

def actividades_criticas():
  "Esta parte se encarga de generar mi archivo txt y que nos escriba la parte de las actividades criticas o las holguras "

  with open('/content/drive/My Drive/Holguras/nuevo.txt', 'w') as file:
        file.write("Las actividades con holgura son: \n")
        file.write("1.-Decision preliminar.\n")
        file.write("2.-Determinacion y analisis de objetivos.\n")
        file.write("3.-Sugerencia del plan de estudios.\n")
        file.write("4.-Presentacion de la propuesta\n")
        file.write("6.-Consulta con el sector academico de la UNAM.\n")
        file.write("11.-Analisis, discusion e incorporacion de sugerencias de la UNAM\n")
        file.write("14.-Analisis comparativo de planes similares en la UNAM.\n")
        file.write("15.-Adicion de consideraciones nacionales.\n")
        file.write("16.-Adicion de consideraciones internacionales.\n1")
        file.write("7.-Elaboracion del documento administrativo.\n")
        file.write("18.-Presentacion en el consejo tecnico de la Facultad de Ciencias.\n")
        file.write("20.-Elaboracion del plan de estudios final.\n")
        file.write("21.-Presentacion de propuesta definitiva a consejos tecnicos y de area.")
        file.write("El tiempo para del proyecto es de 124 semanas. \n")
  file.close()
```
